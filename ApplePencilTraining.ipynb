```python
data_df = pd.DataFrame(columns=["altitude", "azimuth", "force", "x", "y", "speed", "strokeNum"])
# data_df = pd.DataFrame(columns=["altitude", "azimuth", "force", "x", "y", "speed"])
# data_df = pd.DataFrame(columns=["x", "y"])
for d in df:
    data_df.loc[len(data_df.index)]= [d['altitude'].squeeze(), d['azimuth'].squeeze(),d['force'].squeeze(),d['x'].squeeze(),d['y'].squeeze(),d['speed'].squeeze(),d['strokeNum'].squeeze()]
    #      data_df.loc[len(data_df.index)]= [d['altitude'].squeeze(), d['azimuth'].squeeze(),d['force'].squeeze(),d['x'].squeeze(),d['y'].squeeze(),d['speed'].squeeze()]
#     data_df.loc[len(data_df.index)]= [d['x'].squeeze(),d['y'].squeeze()]
    
# print(type(data_df[0]))
print(data_df.dtypes)
# data_df.astype(pd.Series).dtypes
print(data_df['altitude'].dtypes)
data_df.head()
```

```python
X_train = data_df.loc[:71,:]
y_train = name_y[:72]
X_test = data_df.loc[72:,:]
y_test = name_y[72:]
y_train = np.asarray(y_train)
y_test = np.asarray(y_test)
```

```python
from sktime.transformations.panel.rocket import Rocket
from sklearn.linear_model import RidgeClassifierCV
```

```python
from sktime.transformations.panel.padder import PaddingTransformer
```

```python
my_padded_rocket = PaddingTransformer() * RocketClassifier()
my_padded_rocket.fit(X_train, y_train)
```

```python
y_pred = my_padded_rocket.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import precision_score, accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sktime.transformations.panel.rocket import MiniRocketMultivariate
trf = PaddingTransformer() * MiniRocketMultivariate(num_kernels=512) 
```

```python
trf.fit(X_train)
X_train_tr = trf.transform(X_train) 
X_test_tr = trf.transform(X_test) 
```

```python
classifier = RidgeClassifierCV(alphas=np.logspace(-3, 3, 10), normalize=True)
classifier.fit(X_train_tr, y_train)
```

```python
y_pred = classifier.predict(X_test_tr)
print(y_pred)
print(y_test)
accuracy_score(y_test, y_pred)
```
